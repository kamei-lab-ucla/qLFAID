# Imports and Function Definitions

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os
import pandas as pd
from scipy.optimize import least_squares
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.linear_model import LogisticRegression
```

```python
def exponen(theta, x): #exponential
    return theta[0]*np.exp(theta[1]*x)+theta[2]

def log(theta, x): #logarithmic
    return theta[0]*np.log(x + theta[1]) + theta[2]

def inv(theta, x): #inverse
    return (theta[0]/(x+theta[1])) + theta[2]

def poly(theta, x, deg): #polynomial
    out = 0
    for i in range(deg):
        out += theta[i]*(x**(i))
    return out

def lin(theta,x): #linear
    return theta[0]*x + theta[1]

def logistic(x, coef, intercept):
    inp = x * coef.ravel() + intercept
    return 1 / (1 + np.exp(-inp))

def resid_exp(theta,x,y):
    return y-exponen(theta,x)

def resid_log(theta, x, y):
    return y-log(theta,x)

def resid_inv(theta, x, y):
    return y-inv(theta,x)

def resid_lin(theta, x, y):
    return y-lin(theta,x)

def resid_poly(theta, x, deg, y):
    return y-poly(theta,x,deg)
```

```python
#Function for quantifying test line intensity
def testIntensity(src):
    margin = 1 #margin around image in case automated cropping is not perfect, also used when searching for control line location
    scan = 10 #size of area in pixels that is searched for location of test line
    refScan = 10 #size of reference area in pixels
    lineWidth = 6 #Half-width of line
    src = cv2.cvtColor(src, cv2.COLOR_RGB2GRAY) #convert image to grayscale
    img = cv2.rotate(src, cv2.ROTATE_90_COUNTERCLOCKWISE) #rotate image
    
    h,w = img.shape
    
    img = img[margin:h-margin, margin:w-margin]
    
    h,w = img.shape #get height and width
    
    #Convert image to 1D array of pixel intensities. For each row, use median value of grayscale pixel intensity
    mean = np.median(img,axis=0)
    mean = 255 - mean #so that high values correspond to dark areas

    controlToTest = int(w/3) #initial guess at distance between control and test lines
    
    #in case initial guess is too small
    if controlToTest < 100:
        controlToTest = 100
    
    #Looking for control line
    
    #Left/right bounds on area where control line can be found
    left=0 
    right=int(0.4*w)
    
    LFAControl = mean[left:right] #cropped section of area where control line could be
    ControlLineLoc = np.argmax(LFAControl) # potential location of control line
    
    #make sure control line location isn't the leftmost part of the image (this would be inaccurately focusing on a shadow)
    while(ControlLineLoc == left):
        left += margin
        LFAControl = mean[left:right]
        ControlLineLoc = left + np.argmax(LFAControl)
        
    #make sure control line location isn't the rightmost part of the area (this is probably the test line)
    while(ControlLineLoc == right):
        right -= margin
        LFAControl = mean[left:right]
        ControlLineLoc = left + np.argmax(LFAControl)
        
    #Search for test line
    TestLineLoc = ControlLineLoc + controlToTest
    
    if TestLineLoc > len(mean):
        TestLineLoc = ControlLineLoc + 75
    if TestLineLoc + scan > len(mean):
        scan = len(mean)-TestLineLoc
    
    TestLineLocation = TestLineLoc-scan + np.argmax(mean[TestLineLoc-scan:TestLineLoc+scan])
    
    #Determine reference value (value of background itnensity)
    ReferenceLocation = int((ControlLineLoc + TestLineLocation)/2)
    Reference = np.mean(mean[ReferenceLocation-refScan:ReferenceLocation+refScan])
    
    #subtract reference value from entire array, and make sure negative values are converted to 0
    LFACorrected = mean-Reference
    LFACorrected[LFACorrected < 0] = 0
    
    #Determine line intensities by taking sum of corrected intensity, centered at each line location
    ControlLine = np.sum(LFACorrected[ControlLineLoc-lineWidth:ControlLineLoc+lineWidth])
    
    TestLine = np.sum(LFACorrected[TestLineLocation-lineWidth:TestLineLocation+lineWidth])
    
    #return normalized value of test line intensity. Dividing by control line intensity helps account for variations in lighting between pictures + variations in nanoparticle release.
    return TestLine/ControlLine
```

# File Names (please edit)

```python
dir = 'demo_data/quantitative' #folder containing cropped images of detection zones
index_path = 'demo_index.csv' #file name of index file, which identifies the biomarker concentration of each image file name 
test_name = 'LFATest' #name of commercial LFA test
```

Run following cells if interesting in training quantitative calibration curve. If interested in training logistic calibration curve, go to end of document

# Loading Data

```python
#Read index file
index = pd.read_csv(index_path)
SIZE = len(index)
```

```python
#Read image files and calculate test line intensities
conc = np.zeros(SIZE)
intensities = np.zeros(SIZE)
error = np.zeros(5)
param = ['','','','','']

i=0
for filename in os.listdir(dir):
    f = os.path.join(dir, filename)
    img = cv2.imread(f)
    intensities[i] = testIntensity(img)
    conc[i] = (index.loc[index['File Name'] == filename]).values[0,0]
    i += 1
    
min_conc = np.min(conc)
max_conc = np.max(conc)

x_conc = np.linspace(min_conc,max_conc)
```

# Testing Different Regression Models

```python
x0 = [1,1,1] # initial guess at parameters
res_exp = least_squares(resid_exp, x0, args=(intensities, conc)) #calcuate best fit paramters
error[0] = mean_squared_error(conc, exponen(res_exp.x,intensities), squared=False) #determine error
param[0] = res_exp.x #save parameters

#Graph model against data
model = exponen(res_exp.x,x_conc)

plt.plot(x_conc, model, label='Fitted Curve')
plt.scatter(conc,intensities, label='Data')
plt.ylabel("Biomarker Concentration")
plt.xlabel("Normalized Test Line Intensity")
plt.title('Exponential Model')
plt.show()
```

```python
x0 = [1,1,1]
res_log = least_squares(resid_log, x0, args=(intensities, conc))
error[1] = mean_squared_error(conc, log(res_log.x,intensities), squared=False)
param[1] = res_log.x

model = log(res_log.x,x_conc)

plt.plot(x_conc, model, label='Fitted Curve')
plt.scatter(conc,intensities, label='Data')
plt.ylabel("Biomarker Concentration")
plt.xlabel("Normalized Test Line Intensity")
plt.title('Logarithmic Model')
plt.show()
```

```python
x0 = [1,1,1]
res_inv = least_squares(resid_inv, x0, args=(intensities, conc))
error[2] = mean_squared_error(conc, inv(res_inv.x,intensities), squared=False)
param[2] = res_inv.x

model = inv(res_inv.x,x_conc)

plt.plot(x_conc, model, label='Fitted Curve')
plt.scatter(conc,intensities, label='Data')
plt.ylabel("Biomarker Concentration")
plt.xlabel("Normalized Test Line Intensity")
plt.title('Inverse Model')
plt.show()
```

```python
x0 = [1,1]
res_lin = least_squares(resid_lin, x0, args=(intensities, conc))
error[3] = mean_squared_error(conc, lin(res_lin.x,intensities), squared=False)
param[3] = res_lin.x

model = lin(res_lin.x,x_conc)

plt.plot(x_conc, model, label='Fitted Curve')
plt.scatter(conc,intensities, label='Data')
plt.ylabel("Biomarker Concentration")
plt.xlabel("Normalized Test Line Intensity")
plt.title('Linear Model')
plt.show()
```

```python
x0 = [1,1,1]
res_quad = least_squares(resid_poly, x0, args=(intensities, 2, conc))
error[4] = mean_squared_error(conc, poly(res_quad.x,intensities, 2), squared=False)
param[4] = res_quad.x

model = poly(res_quad.x,x_conc,2)

plt.plot(x_conc, model, label='Fitted Curve')
plt.scatter(conc,intensities, label='Data')
plt.ylabel("Biomarker Concentration")
plt.xlabel("Normalized Test Line Intensity")
plt.title('Quadratic Model')
plt.show()
```

# Determining Best Model + Outputting Parameters

```python
models = ['exponential', 'logarithmic', 'inverse', 'linear', 'quadratic']
min_i = np.argmin(error) #Best model is the model with the lowest root-mean-squared error

path = r'test_parameters/' + test_name + '.txt'

#Writing output to text file
f = open(path, 'a')
print('Test Name: ' + test_name, file=f)
print('Model Type: ' + models[min_i], file=f)
print('Parameters: ' + str(param[min_i]), file=f)
f.close()

#Printing output
print('Test Name: ' + test_name)
print('Model Type: ' + models[min_i])
print('Parameters: ' + str(param[min_i]))
```

# Training Logistic Model

```python
neg_dir = r"C:\Users\rohan\OneDrive\Documents\GitHub\qLFAID\demo_data\negative"
pos_dir = r"C:\Users\rohan\OneDrive\Documents\GitHub\qLFAID\demo_data\positive"

# We recommend at least 3 separate pictures in each directory
```

```python
# Count files
count = 0
for path in os.listdir(neg_dir):
    if os.path.isfile(os.path.join(neg_dir, path)):
        count += 1
for path in os.listdir(pos_dir):
    if os.path.isfile(os.path.join(pos_dir, path)):
        count += 1
LOG_SIZE = count
```

```python
x = np.zeros(LOG_SIZE)
y = np.zeros(LOG_SIZE)

i=0

for filename in os.listdir(neg_dir):
    f = os.path.join(neg_dir, filename)
    img = cv2.imread(f)
    x[i] = testIntensity(img)
    y[i] = 0
    i += 1
    
for filename in os.listdir(pos_dir):
    f = os.path.join(pos_dir, filename)
    img = cv2.imread(f)
    x[i] = testIntensity(img)
    y[i] = 1
    i += 1
        
classifier = LogisticRegression(random_state = 0)
classifier.fit(x.reshape(-1,1),y)

plt.scatter(x, y, label='Data')
int_test = np.linspace(np.min(x),np.max(x))
loss = logistic(int_test,classifier.coef_,classifier.intercept_).ravel()
plt.plot(int_test, loss, label='Fitted Curve')
plt.ylabel("Test Result")
plt.xlabel("Normalized Test Line Intensity")
plt.title('Logistic Model')
plt.legend()
plt.show()
```

```python
path = r'test_parameters/' + test_name + '.txt'

#Writing output to text file
f = open(path, 'a')
print('Test Name: ' + test_name, file=f)
print('Model Type: Logistic', file=f)
print('Coefficient: ' + str(classifier.coef_[0,0]), file=f)
print('Intercept: ' + str(classifier.intercept_[0]), file=f)
f.close()

#Printing output
print('Test Name: ' + test_name)
print('Model Type: ' + models[min_i])
print('Coefficient: ' + str(classifier.coef_[0,0]))
print('Intercept: ' + str(classifier.intercept_[0]))
```
